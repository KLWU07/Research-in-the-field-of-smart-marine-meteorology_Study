```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 
data = pd.read_csv(r'C:\Users\11712\notebook\Untitled Folder\train\train_ID01.csv')

#  'date_time'  'wind_speed' 
data['date_time'] = pd.to_datetime(data['date_time'])  #  datetime 
data.set_index('date_time', inplace=True)  # 

# 2023060120240630
start_date = '2023-06-01 00:00:00'
end_date = '2023-06-30 23:45:00'
filtered_data = data[(data.index >= start_date) & (data.index <= end_date)]

# 
plt.figure(figsize=(15, 6))  # 
plt.plot(filtered_data.index, filtered_data['wind_obs'], label='Wind Speed', color='blue')  # 

# 
# 
time_index = np.arange(len(filtered_data.index))

# numpypolyfit
coefficients = np.polyfit(time_index, filtered_data['wind_obs'], deg=1)  # deg=1
polynomial = np.poly1d(coefficients)  # 
trend_line = polynomial(time_index)  # 

# 
plt.plot(filtered_data.index, trend_line, label='Trend Line', color='red', linestyle='--')

# 
plt.title('Wind Speed from 2023-06-01 to 2023-06-15', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Wind Speed', fontsize=12)

# 
plt.legend(fontsize=10)

# x
plt.gcf().autofmt_xdate()  # 
plt.xticks(rotation=45)  # x

# 
plt.tight_layout()  # 
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 
data = pd.read_csv(r'C:\Users\11712\notebook\Untitled Folder\train\train-11.csv')

#  'date_time' 
data['date_time'] = pd.to_datetime(data['date_time'])  #  datetime 
data.set_index('date_time', inplace=True)  # 

# 2023060120240630
start_date = '2023-06-01 00:00:00'
end_date = '2024-05-31 23:45:00'
filtered_data = data[(data.index >= start_date) & (data.index <= end_date)]

# 'date_time'
wind_speed_columns = [col for col in filtered_data.columns if col != 'date_time']

# 3x4
fig, axes = plt.subplots(3, 4, figsize=(20, 15))  # 3x4
axes = axes.flatten()  # axes

# 
for i, column in enumerate(wind_speed_columns):
    if i >= 12:  # 123x4
        break
    
    # 
    wind_speed = filtered_data[column]
    
    # 
    axes[i].plot(wind_speed.index, wind_speed, label=f'{column} Wind Speed', color='blue')
    
    # 
    time_index = np.arange(len(wind_speed.index))
    coefficients = np.polyfit(time_index, wind_speed, deg=1)  # 
    polynomial = np.poly1d(coefficients)
    trend_line = polynomial(time_index)
    axes[i].plot(wind_speed.index, trend_line, label='Trend Line', color='red', linestyle='--', linewidth=2)
    
    # 
    axes[i].set_title(f'{column} Wind Speed', fontsize=12)
    axes[i].set_xlabel('Date', fontsize=10)
    axes[i].set_ylabel('Wind Speed', fontsize=10)
    axes[i].legend(fontsize=8)
    axes[i].tick_params(axis='x', rotation=45)  # x

# 
for j in range(len(wind_speed_columns), 12):
    fig.delaxes(axes[j])

# 
plt.tight_layout()
plt.show()
```
