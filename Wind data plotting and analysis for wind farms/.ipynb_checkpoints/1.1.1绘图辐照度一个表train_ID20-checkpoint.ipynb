```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 
data = pd.read_csv(r'train/train_ID20.csv')

#  'date_time'  'wind_speed' 
data['date_time'] = pd.to_datetime(data['date_time'])  #  datetime 
data.set_index('date_time', inplace=True)  # 

# 2023060120240630
start_date = '2023-06-01 00:00:00'
end_date = '2024-06-30 23:45:00'
filtered_data = data[(data.index >= start_date) & (data.index <= end_date)]

# 
plt.figure(figsize=(15, 6))  # 
plt.plot(filtered_data.index, filtered_data['fuzhaodu_obs'], label='Wind Speed', color='blue')  # 

# 
plt.title('Wind Speed from 2023-06-01 to 2024-06-30', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Wind Speed', fontsize=12)

# 
plt.legend(fontsize=10)

# x
plt.gcf().autofmt_xdate()  # 
plt.xticks(rotation=45)  # x

# 
plt.tight_layout()  # 
plt.show()
```
