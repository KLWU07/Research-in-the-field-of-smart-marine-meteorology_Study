# CSV表中读取两列ec、wind_obs计算皮尔逊相关系数，筛选时间从2023.07.02到2024.06.30


```python
import pandas as pd
from scipy.stats import pearsonr

# 读取CSV文件
file_path = r'E:\我的桌面\03.01之前风力发电代码\1ECMWF预报数据和11风场观测数据训练模型\ID01.csv'  # 替换为你的CSV文件路径
data = pd.read_csv(file_path)

# 确保时间列是datetime类型（假设时间列名为 'time'）
data['date_time'] = pd.to_datetime(data['date_time'])

# 筛选时间范围：2023.07.02 到 2024.06.30
start_date = pd.to_datetime('2023-07-01 00:00:00')
end_date = pd.to_datetime('2024-06-30 18:00:00')
filtered_data = data[(data['date_time'] >= start_date) & (data['date_time'] <= end_date)]

# 提取需要的两列：ec 和 wind_obs
ec = filtered_data['ec']
wind_obs = filtered_data['wind_obs']

# 计算皮尔逊相关系数
corr, p_value = pearsonr(ec, wind_obs)

# 输出结果
print(f"皮尔逊相关系数: {corr:.4f}")
print(f"P值: {p_value:.4f}")
```
