```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 导入数据
data = pd.read_csv(r'C:\Users\11712\notebook所有\Untitled Folder\train\train-10.csv')

# 确保时间列是 datetime 类型，并设置为索引
data['date_time'] = pd.to_datetime(data['date_time'])
data.set_index('date_time', inplace=True)

# 筛选出2023年06月01日到2024年06月30日的数据
start_date = '2023-06-01 00:00:00'
end_date = '2024-05-31 23:45:00'
filtered_data = data[(data.index >= start_date) & (data.index <= end_date)]

# 绘制箱线图
plt.figure(figsize=(14, 8))  # 设置图形大小
sns.boxplot(data=filtered_data)  # 使用 seaborn 绘制箱线图

# 添加标题和标签
plt.title('Wind Speed Distribution (2023-06-01 to 2024-05-31)', fontsize=14)
plt.xlabel('Wind Field', fontsize=12)
plt.ylabel('Wind Speed', fontsize=12)
plt.xticks(rotation=45)  # 旋转 x 轴标签，避免重叠

# 显示图形
plt.tight_layout()  # 自动调整布局
plt.show()
```

```python
#箱线图设置指定的颜色列表
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 导入数据
data = pd.read_csv(r'C:\Users\11712\notebook所有\Untitled Folder\train\train-10.csv')

# 确保时间列是 datetime 类型，并设置为索引
data['date_time'] = pd.to_datetime(data['date_time'])
data.set_index('date_time', inplace=True)

# 筛选出2023年06月01日到2024年06月30日的数据
start_date = '2023-06-01 00:00:00'
end_date = '2024-05-31 23:45:00'
filtered_data = data[(data.index >= start_date) & (data.index <= end_date)]

# 定义颜色列表
colors = ['springgreen', 'green', 'red', 'cyan', 'yellow', 'greenyellow', 'orange', 'purple', 'brown', 'pink', 'gray']

# 绘制箱线图
plt.figure(figsize=(14, 8))  # 设置图形大小
sns.boxplot(data=filtered_data, palette=colors)  # 使用 seaborn 绘制箱线图，并设置颜色

# 添加标题和标签
plt.title('Wind Speed Distribution (2023-06-01 to 2024-05-31)', fontsize=14)
plt.xlabel('Wind Field', fontsize=12)
plt.ylabel('Wind Speed', fontsize=12)
plt.xticks(rotation=45)  # 旋转 x 轴标签，避免重叠

# 自动调整布局
plt.tight_layout()

# 保存图片
output_path = r'C:\Users\11712\notebook所有\Untitled Folder\wind_speed_boxplot.png'  # 定义保存路径
plt.savefig(output_path, dpi=300)  # 保存图片，设置分辨率为300dpi

# 显示图形
plt.show()
```
