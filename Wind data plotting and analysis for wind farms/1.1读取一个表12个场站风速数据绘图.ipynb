```python
# 2023.06.01-2024.06.30
import pandas as pd
import matplotlib.pyplot as plt

# 
data = pd.read_csv(r'C:\Users\11712\notebook\Untitled Folder\train\train-12.csv')

#  'date_time' 
data['date_time'] = pd.to_datetime(data['date_time'])  #  datetime 
data.set_index('date_time', inplace=True)  # 

# 2023060120240630
start_date = '2023-06-01 00:00:00'
end_date = '2024-06-30 23:45:00'
filtered_data = data[(data.index >= start_date) & (data.index <= end_date)]

#  'wind_speed_1', 'wind_speed_2', 'wind_speed_3'
columns_to_plot = ['wind_obs1', 'wind_obs2', 'wind_obs3', 'wind_obs7']

# 
plt.figure(figsize=(15, 6))  # 

# 
for column in columns_to_plot:
    plt.plot(filtered_data.index, filtered_data[column], label=column)

# 
plt.title('Wind Speed from 2023-06-01 to 2024-06-30 for Selected Wind Farms', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Wind Speed', fontsize=12)

# 
plt.legend(fontsize=10)

# x
plt.gcf().autofmt_xdate()  # 
plt.xticks(rotation=45)  # x

# 
plt.tight_layout()  # 
plt.show()
```
